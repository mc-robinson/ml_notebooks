# Lagrange Interpolation #

The Lagrange interpolant is defined as follows:

$$
p ( x ) = \sum _ { k = 0 } ^ { n }y_ { k } L _ { k } ( x ) = \sum _ { k = 0 } ^ { n } y _ { k } \prod _ { l = 0 \atop l \neq k } ^ { n } \frac { x - x _ { l } } { x _ { k } - x _ { l } }
$$

where the $L_k$ are the *Lagrange Cardinal Polynomials*.

For example, with $n=2$, the interpolant is a quadratic polynomial $P_2(x) = y _ { 0 } L _ { 0 } ( x ) + y _ { 1 } L _ { 1 } ( x ) + y _ { 2 } L _ { 2 } ( x )$, where the respective cardinal functions are defined as:

$$
\left.\begin{aligned} L _ { 0 } ( x ) & = \frac { \left( x - x _ { 1 } \right) \left( x - x _ { 2 } \right) } { \left( x _ { 0 } - x _ { 1 } \right) \left( x _ { 0 } - x _ { 2 } \right) } \\ L _ { 1 } ( x ) & = \frac { \left( x - x _ { 0 } \right) \left( x - x _ { 2 } \right) } { \left( x _ { 1 } - x _ { 0 } \right) \left( x _ { 1 } - x _ { 2 } \right) } \\ L _ { 2 } ( x ) & = \frac { \left( x - x _ { 0 } \right) \left( x - x _ { 1 } \right) } { \left( x _ { 2 } - x _ { 0 } \right) \left( x _ { 2 } - x _ { 1 } \right) } \end{aligned} \right.
$$

Looking at the form of the cardinal functions, one can see the following property:

$$
L _ { i } \left( x _ { j } \right) = \left\{ \begin{array} { l } { 0 \text { if } i \neq j } \\ { 1 \text { if } i = j } \end{array} \right\} = \delta _ { i j }
$$

Thus, each individual cardinal polynomial $L_k(x)$ will equal $y_k=f(x_k)$ at $x_k$, but will have a root (equal 0) at other data points $x_j$. It is also quite easy to prove that the overall interpolating polynomial passes through every data point $(x_j, y_j)$. The proof is quite short:

$$
P _ { n } ( x _ { j }) = \sum _ { i = 0 } ^ { n } y _ { i } L _ { i }( x _ { j }) = \sum _ { i = 0 } ^ { n } y _ { i } \delta _ { i j } = y _ { j }
$$

Let's now code up a function to make the interpolating Lagrange polynomial given a set of $x$ and $y$ data points.


```python
def lagrange(x_values, y_values):
    '''
    Find Lagrange interpolant given lists
    of (x,y) data points
    
    x_values, y_values are lists of values in the style of matplotlib
    '''
        
    n = len(x_values)
    if len(y_values) != n:
        raise ValueError("input lists must be of same length")

    # list to hold all of the cardinal polynomials
    Li_list = []
    for i in range(0,n):
        def Li(x,i):
            cardinal_total = 1
            for l in range(0,n):
                if l == i:
                    continue
                cardinal_total *= (x-x_values[l])/(x_values[i]-x_values[l])
            return cardinal_total
        Li_list.append(Li)
    
    # return the polynomial P(x),
    # by summing up all the Li*yi
    def P(x):
        P_total = 0
        for i, Li in enumerate(Li_list):
            P_total += y_values[i]*Li_list[i](x,i)
        return P_total
        
    return P, Li_list
```
